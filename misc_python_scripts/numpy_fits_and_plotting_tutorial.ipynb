```python
from astropy.io import fits

path = '/home/gmbrandt/Documents/Repositories/SRA2021_Track3/lab1_resources/ogg0m406-kb27-20210612-0230-s91.fits.fz'

# opening a file in fits. This gives a list of "hdu"s
hdu_list = fits.open(path)
print(type(hdu_list[0]))
print(hdu_list)
```

```python
# get info about the observation
print(hdu_list.info())

sci_image = hdu_list['SCI'].data
sci_header = hdu_list['SCI'].header
```

```python
sci_header
```

```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

# plt.figure() initializes this "figure"
plt.figure()

# after figure, you do your plotting command
plt.imshow(sci_image)

# plt.show() tells python to show the plot.
plt.show()
```

```python
# plt.figure() initializes this "figure"
plt.figure()

# after figure, you do your plotting command
plt.imshow(sci_image, cmap='Greys_r')

# plt.show() tells python to show the plot.
plt.show()
```

```python
from matplotlib.colors import LogNorm

%matplotlib notebook

# plt.figure() initializes this "figure"
plt.figure()

# after figure, you do your plotting command
plt.imshow(sci_image, norm=LogNorm(vmin=1, vmax=30000), cmap='Greys_r')

# plt.show() tells python to show the plot.
plt.show()
```

```python
%matplotlib inline
# plt.figure() initializes this "figure"
plt.figure(figsize=(13, 13))

# after figure, you do your plotting command
plt.imshow(sci_image[500:1000, 700:1100], norm=LogNorm(vmin=1, vmax=30000), cmap='Greys_r')
plt.gca().invert_yaxis()

# plt.show() tells python to show the plot.
plt.show()
```
